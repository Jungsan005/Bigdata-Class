```python
import os
import pandas as pd
import datetime
```

```python
path1 = 'E:/BD_Class/project/02. 아파트 전월세'
path2 = 'E:/BD_Class/project/04. 연립다세대 전월세'
path3 = 'E:/BD_Class/project/06. 단독다가구 전월세'
p_list1 = os.listdir(path1)
p_list2 = os.listdir(path2)
p_list3 = os.listdir(path3)
```

```python
import shutil
```

```python
for file in p_list1:
    source_file = path1+'/'+file
    destination_path = 'E:/BD_Class/project/전월세'
    shutil.copy2(source_file,destination_path)
for file in p_list2:
    source_file =  path2+'/'+file
    destination_path = 'E:/BD_Class/project/전월세'
    shutil.copy2(source_file,destination_path)
for file in p_list3:
    source_file =  path3+'/'+file
    destination_path = 'E:/BD_Class/project/전월세'
    shutil.copy2(source_file,destination_path)
p_list = p_list1+p_list2+p_list3
gu_list = []
for gu in p_list:
    gu_list.append(gu.split('(')[0].split('.')[1][1:])

# dt_list = [i[-13:-7] for i in p_list]
print(p_list,gu_list)
```

```python
pd.read_csv(destination_path+'/'+'01. 종로구(단독다가구 전월세)_2010_2019.csv',encoding='euc-kr')
```

```python
dic = dict()
for gu, gu_file in zip(gu_list,p_list):
    dic[gu]=pd.read_csv(destination_path+'/'+gu_file,encoding='euc-kr')
```

```python
dic['종로구']
```

```python
temp = []
for j in gu_list:
    temp.append(dic[j])
```

```python
df = pd.concat(temp)
```
