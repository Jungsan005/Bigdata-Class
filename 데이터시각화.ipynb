```
# 정규분포를 따르는 임의의 난수 15개 생성
# 정규분포 : 수집한 데이터의 분포유형이 종모양, 평균은 0, 표준편차 1인 형태
```

```
x <- seq(-3,3,length=200)    # 임의의 난수 생성
```

```
y <- rnorm(50,0,1) # 갯수, 평균, 표준편차
head(y)
```

```
plot(x, dnorm(x, 0, 1), type="l")  # 평균이 0이고, 표준편차가 1인 정규분포 (밀도함수)
```

```
barplot(y) 
```

```
stem(y) # 줄기-잎 그래프
```

```
# 그래프의 다양한 설정하기
# main, sub, xlab/ylab, axes, col, pch
```

```
plot(heights, main="student's height", sub="unit:cm", col.main="red", col.sub="blue", xlab="std no.", ylab = "height")
```

```
# 축 간격 조정 : xlim, ylim
plot(heights, xlim = c(0,12), ylim = c(145,185))
```

```
plot(heights, xlim = c(0,12), ylim = c(145,185), las=1)
```

```
# x,y축 사용자 지정: axes(x축/y축, 범위)
plot(heights, axes=F) # x축/y축 제외하고 그림
axis(1, at=1:10)
axis(2, at=seq(140, 185, 5), las=1)
box()
```

```
# 각 점에 수치 표시
# text(값, 굵기, 표시문자, 위치)
# 위치: adj()
# 위치지정값: 음수: 오른쪽, 위, 양수: 왼쪽, 아래

plot(heights, col="red")
text(heights, cex=0.8, lab=heights, adj=c(-0,1))
```
